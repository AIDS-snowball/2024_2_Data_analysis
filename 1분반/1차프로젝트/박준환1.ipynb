```python
import pandas as pd
df= pd.read_csv ('/train.csv', index_col=0)
selected=df.drop(['Name','Sex','Ticket','Fare','Embarked'],axis=1)

print("총 생존율:",selected['Survived'].mean())
print('\n')

print("객실 등급별 생존율")
print(selected.groupby('Pclass')['Survived'].mean())
print('\n')

print("가족 수별 생존율")
print(selected.groupby('SibSp')['Survived'].mean())
print('\n')

print("부모 자녀 수별 생존율")
print(selected.groupby('Parch')['Survived'].mean())
```
