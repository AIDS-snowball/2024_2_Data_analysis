# 개요

상황에 딱 맞는 데이터를 원하는 만큼 구하기는 쉽지 않음

그러므로 공개 데이터 세트를 찾아보거나 온라임 포럼에 도움을 요청할 수 있음

* 공공 데이터 포털
* 도서관 정보나루

````python
import gdown

gdown.download('https://bit.ly/3eecMKZ', '남산도서관 장서 대출목록 (2021년 04월 ).csv', quiet=False)
````

> gdown 패키지는 구글 드라이브를 포함하여 웹에서 대용량 파일을 다운로드할 수 있는 패키지이다.



````python
with open('남산도서관 장서 대출목록 (2021년 04월 ).csv') as f:
  print(f.readline())
````

> **UnicodeDecodeError:**<br />
open() 함수는 기본적으로 UTF-8형식으로 저장<br />
한글 텍스트는 EUC-KR을 사용하는 일이 잦음

````python
import chardet
with open("남산도서관 장서 대출목록 (2021년 04월 ).csv", mode='rb') as f:
  d = f.readline()
print(chardet.detect(d))
````

> **chardet, chardet.detect()함수 :** <br/>
문자 인코딩 방식을 알아낼 수 있음

> **open()함수의 매개변수 'rd':**<br/>
문자 인코딩 형식에 상관없이 파일을 열 수 있음

````python
with open('남산도서관 장서 대출목록 (2021년 04월 ).csv', encoding="EUC-KR:") as f:
  print(f .readline())
  print(f .readline())
````

CSV 파일은 판다스와 같은 도구로 읽는 것이 조금 더 편리

아주 큰 파일을 열 때는 오랜 시간이 걸리거나 아예 열수 없을 떄도 있음

파이썬의 open()함수와 readline() 메서드로 청므 몇 줄을 출력해 보는 거이 빠르게 내용을 확인할 수 있는 방법 중 하나

> 한글을 유니코드로 표현할 때 글자 단위로 저장하거나 자음 모음을 따로따로 저장할 수 있습니다. 예를 들어 혼공을 ‘ㅎㅗㄱㅗㅇ’처럼 저장할 수 있죠. 전자를 NFC 방식이라 하고 후자를 NFD 방식이라 합니다. 이렇게 두 가지 방식이 있는 이 유는 유니코드에 한글의 자음, 모음 코드가 별도로 있기 때문입니다.<br/>
윈도우와 리눅스는 NFC 방식을 사용하지만 맥OS는 NFD 방식을 사용합니다. 이 때문에 맥OS에서 만든 파일을 윈도우 에서 보면 파일 이름의 자음 모음이 모두 분리되어 보이게 됩니다. 만약 맥OS 사용자에게 파일을 받았거나 NFD 방식의 이름이라고 의심된다면 다음 코드를 실행하여 NFC 방식의 이름으로 바꿀 수 있습니다.

```
import os
import glob
import unicodedata
for filename in glob.glob('*.csv'):
nfc_filename = unicodedata.normalize('NFC', filename) os.rename(filename, nfc_filename)
```
