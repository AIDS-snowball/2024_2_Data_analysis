# matplotlib의 객체

## figure, rcParams, subplot

# Figure

다양한 그래프의 옵션을 조절할 수 있다.

```python
import gdown

gdown.download('https://bit.ly/3pK7iuu', 'ns_book7.csv', quiet=False)
```

```python
import pandas as pd

ns_book7 = pd.read_csv('ns_book7.csv', low_memory=False)
ns_book7.head()
```

```python
import matplotlib.pyplot as plt

plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

## figsize
그래프 크기 바꾸기

```python
plt.figure(figsize=(9,6)) # 너비 9인치, 높이 6인치의 피겨 객체
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

### 그래프 크기 확인


```python
print(plt.rcParams['figure.figsize'])
```

### DPI값 확인
프로그램마다 차이 존재

```python
print(plt.rcParams['figure.dpi'])
```

### 900*600픽셀 크기의 그래프 그리기

```python
plt.figure(figsize=(900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

## 주변 공백을 최소화하는 타이츠 레이아웃
기본적으로 적용되어 위의 그래프가 실제 900*600이 아니게 됨

```python
%config InLineBackend.print_figure_kwargs = {'bbox_inches':None}
plt.figure(figsize=(900/72, 600/72))
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

## dpi 매개변수, dpi 변경

```python
plt.figure(dpi=144)
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha = 0.1)
plt.show()
```

# reParams
맥플롯립 그래프의 기본값을 관리하는 객체

## DPI 변셩

```python
plt.rcParams['figure.dpi'] = 100
```

```python
plt.rcParams['scatter.marker']
```

```python
plt.rcParams['scatter.marker'] = '*'
```

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
plt.show()
```

### 걍 바꾸기

```python
plt.scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1, marker="+")
plt.show()
```

# subplot
Axes 클래스 : 축
tick : 각 축의 눈금
label : 축의 이름image.png

```python
fig, axs = plt.subplots(2)

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_yscale('log')

fig.show()
```

## figsize
피겨 크기를 지정

## set_title()

```python
fig, axs = plt.subplots(2, figsize=(6,8))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')

fig.show()
```

## 가로로 출력하기

## 레이블 이름 지정하기
set_xlabel(), set_ylabel()

```python
fig, axs = plt.subplots(1,2, figsize=(10,4))

axs[0].scatter(ns_book7['도서권수'], ns_book7['대출건수'], alpha=0.1)
axs[0].set_title('scatter plot')
axs[0].set_xlabel('number of books')
axs[0].set_ylabel('borrow count')

axs[1].hist(ns_book7['대출건수'], bins=100)
axs[1].set_title('histogram')
axs[1].set_yscale('log')
axs[1].set_xlabel('borrow count')
axs[1].set_ylabel('frequency')

fig.show()
```
